# Descrição do Problema

Esse conjunto de dados aborda o desempenho escolar de estudantes do ensino secundário em duas escolas portuguesas. Os atributos dos dados incluem notas dos alunos, características demográficas, sociais e relacionadas à escola, e foram coletados por meio de boletins escolares e questionários.

No estudo de Cortez e Silva (2008), esses dados foram modelados em tarefas de classificação binária, classificação em cinco níveis e regressão.

## Hipóteses do Problema

As seguintes hipóteses foram consideradas:

* Alunos que têm mais tempo de estudo por semana (TempoEstudo) tendem a ter notas mais altas?

* Alunos com algum tipo de apoio educacional (ApoioFamilia, ApoioEscola ou ApoioExtra) possuem notas mais altas?

## Tipos de Problema
Este é um problema de **classificação supervisionada.**
Dado um conjunto de características dos alunos — como tempo de estudo semanal, tipo e quantidade de apoio educacional recebido (familiar, escolar ou externo), escolaridade dos pais e outras variáveis socioeducacionais — o objetivo é prever se existe alguma correlação entre o tempo de estudo e os tipos de apoio no desempenho acadêmico dos alunos.


## Seleção de Dados

Foi realizado um merge dos arquivos originais "student-mat.csv" e "student-por.csv", gerando o arquivo "DesempenhoEstudantes.csv" que será utilizado neste MVP.

Todas as colunas foram traduzidas para português para facilitar o entendimento, e algumas colunas já possuem técnicas de classificação e regressão aplicadas.

## Total e Tipo das Instâncias
O dataset possui 1044 instâncias, sendo divididas entra as matérias Português e Matemática. Vamos analisar o tempo de estudo, o tipo de apoio educacional, a localização da moradia e a escolaridade dos pais de cada aluno.

```python
print(f"Total de instâncias: {len(df)}")
```

```python
# --- Função para configurar eixos ---
def configurar_eixos(ax, titulo='', xlabel='', ylabel=''):
    ax.set_title(titulo)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

# --- Função para gerar os rótulos abreviados das combinações de apoio educacional ---
def gerar_rotulos_apoios(df):
    agrupado_apoios = df.groupby(['ApoioExtra', 'ApoioFamilia', 'ApoioEscola']).size().reset_index(name='frequencia')
    agrupado_apoios['rotulo'] = (
        'Extra:' + agrupado_apoios['ApoioExtra'].str[0].str.upper() + '\n' +
        'Familiar:' + agrupado_apoios['ApoioFamilia'].str[0].str.upper() + '\n' +
        'Escolar:' + agrupado_apoios['ApoioEscola'].str[0].str.upper()
    )
    return agrupado_apoios

# --- Função para gerar os rótulos abreviados das combinações de escolaridade dos pais ---
def gerar_rotulos_escolaridades(df, col_mae, col_pai, rotulos_dict):
    df_tmp = df.copy()
    df_tmp['MaeRotulo'] = df_tmp[col_mae].map(rotulos_dict)
    df_tmp['PaiRotulo'] = df_tmp[col_pai].map(rotulos_dict)
    agrupado_escolaridade = df_tmp.groupby(['MaeRotulo', 'PaiRotulo']).size().reset_index(name='frequencia')
    agrupado_escolaridade['RotuloCombinado'] = 'Mãe: ' + agrupado_escolaridade['MaeRotulo'] + '\nPai: ' + agrupado_escolaridade['PaiRotulo']
    return agrupado_escolaridade

# --- Função para gerar os rótulos abreviados das combinações de escolaridade dos pais ---
def gerar_rotulos_escolaridade(row):
    mae = row['MaeEscolaridade']
    pai = row['PaiEscolaridade']

    if mae == 0 and pai == 0:
        return 'Nenhum'

    maior = max(mae, pai)

    if maior <= 2:
        return 'Fundamental'
    elif maior == 3:
        return 'Médio'
    else:
        return 'Superior'

df['EscolaridadePaisNivel'] = df.apply(gerar_rotulos_escolaridade, axis=1)

# --- Função para adicionar rótulos acima das barras ---
def adicionar_valores(grf, deslocamento=0.2):
    for barra in grf.patches:
        altura = barra.get_height()
        grf.text(barra.get_x() + barra.get_width()/2, altura + deslocamento,
                 f'{int(altura)}', ha='center', va='bottom')

# --- Dicionário de rótulos ---
# Tempo de Estudo
rotulos_estudo = {
    1: '<2 horas',
    2: '2 a 5 horas',
    3: '5 a 10 horas',
    4: '>10 horas'
}

df['TempoEstudoDescricao'] = df['TempoEstudo'].map(rotulos_estudo)
ordem_estudo = [rotulos_estudo[i] for i in sorted(rotulos_estudo)]

# Endereço (Localização)
rotulos_localizacao = {
    'U': 'Urbana',
    'R': 'Rural'
}

df['EnderecoDescricao'] = df['Endereco'].map(rotulos_localizacao)

# --- Dados das combinações ---
df_apoios = gerar_rotulos_apoios(df)

df_escolaridade = df['EscolaridadePaisNivel'].value_counts().reset_index()
df_escolaridade.columns = ['nivel', 'frequencia']

# --- Criação da figura com 2 linhas e 2 colunas de gráficos ---
fig, axes = plt.subplots(2, 2, figsize=(20, 14))

# facilita acesso como lista
axes = axes.flatten()

# --- Gráfico 1: Tempo de Estudo ---
sns.countplot(x='TempoEstudoDescricao', data=df, ax=axes[0], order=ordem_estudo)
configurar_eixos(axes[0], titulo='Tempo de Estudo')
adicionar_valores(axes[0])

# --- Gráfico 2: Apoio Educacional (Combinações) ---
sns.barplot(x='rotulo', y='frequencia', data=df_apoios, ax=axes[1])
configurar_eixos(axes[1], titulo='Apoio Educacional (Combinações)')
adicionar_valores(axes[1], deslocamento=0.3)

for label in axes[1].get_xticklabels():
    label.set_ha('center')

# --- Gráfico 3: Localização ---
sns.countplot(x='EnderecoDescricao', data=df, ax=axes[2])
configurar_eixos(axes[2], titulo='Endereço (Localização)')
adicionar_valores(axes[2])

# --- Gráfico 4: Escolaridade dos Pais (Combinações) ---
sns.barplot(x='nivel', y='frequencia', data=df_escolaridade, ax=axes[3], order=['Nenhum', 'Fundamental', 'Médio', 'Superior'])
configurar_eixos(axes[3], titulo='Escolaridade dos Pais (Combinações)')
adicionar_valores(axes[3])

# --- Plotagem ---
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.15)
plt.show()
```

- Os gráficos mostram que todas as situações são desbalanceadas
- No gráfico "Apoio Educacional" foram exibidas todas as combinaçoes
- No gráfico "Escolaridade dos Pais" foi considerada a **maior** escolaridade entre os pais

# Analisando o Tempo de Estudo

## Correlação

```python
# --- Correlação entre o tempo de estudo e a nota final ---
df_grupo = df.groupby('TempoEstudoDescricao')['P3'].mean().reindex(ordem_estudo)

# --- Print ---
print(df_grupo.reset_index())
```

- Tendência geral de crescimento
  - À medida que o tempo de estudo aumenta, a média da nota também sobe.
  - Isso sugere uma correlação positiva entre tempo de estudo e desempenho acadêmico.
- Ponto de inflexão após 10 horas
  - A média cai levemente de 12.49 para 12.27 no grupo que estuda mais de 10 horas.
  - Isso pode indicar:
    - Um limite de retorno: estudar mais que 10h/semana não garante melhora adicional
    - Ou que esse grupo inclui alunos com dificuldades que estudam mais, mas ainda enfrentam desafios
- Impacto real do estudo
  - A diferença entre o grupo que estuda <2h e o que estuda 5–10h é de quase 2 pontos na média final
  - Isso é significativo e reforça que tempo de estudo é um fator relevante no desempenho


## Boxplot

```python
# --- Criação da figura ---
fig, ax = plt.subplots(figsize=(8, 5))

# --- Gráfico ---
sns.boxplot(x='TempoEstudoDescricao', y='P3', data=df, order=ordem_estudo)
configurar_eixos(ax, titulo='Distribuição das Notas Finais (P3) por Tempo de Estudo')

# --- Plotagem ---
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.15)
plt.show()
```

- Crescimento até 10 horas
  - A mediana da nota aumenta progressivamente do grupo que estuda menos de 2h até o grupo de 5 a 10h.
  - Isso reforça a ideia de que estudar mais está associado a melhor desempenho, até certo ponto.
- Leve recuo após 10h
  - O grupo que estuda mais de 10h/semana apresenta uma mediana levemente inferior ao grupo de 5 a 10h.
  - Isso pode indicar:
    - Diminuição do retorno marginal: estudar mais não garante melhora contínua
    - Ou que esse grupo inclui alunos com dificuldades que estudam mais para compensar
- Dispersão semelhante entre os grupos
  - A largura das caixas (intervalo interquartil) é parecida em todos os grupos.
  - Isso sugere que a variabilidade do desempenho é constante, independentemente do tempo de estudo.
- Presença de outliers em todos os grupos
  - Há alunos com notas muito baixas (<5) em todas as faixas de estudo.
  - Isso mostra que tempo de estudo não é o único fator determinante — outros elementos como apoio, ambiente, motivação e base anterior também influenciam.

# Analisando o Tipo de Apoio Educacional

## Boxplot

```python
# --- Criar rótulo único das combinações de apoio ---
df['CombinacaoApoio'] = (
    'Extra: ' + df['ApoioExtra'] + '\n' +
    'Família: ' + df['ApoioFamilia'] + '\n' +
    'Escola: ' + df['ApoioEscola']
)

# --- Ordena as combinações pela frequência ---
ordem_combinacoes = df['CombinacaoApoio'].value_counts().index.tolist()

# --- Criação da figura ---
fig, ax = plt.subplots(figsize=(12, 6))

# --- Gráfico boxplot das notas por combinação de apoio ---
sns.boxplot(x='CombinacaoApoio', y='P3', data=df, order=ordem_combinacoes)
configurar_eixos(ax, titulo='Distribuição da Nota Final (P3) por Combinação de Apoios')

# --- Plotagem ---
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.15)
plt.show()
```

- Alguns grupos apresentam:
    - Medianas visivelmente mais altas (por exemplo, quem recebe múltiplos apoios)
    - Outros mostram alta dispersão e notas mais baixas (notavelmente o grupo "Sem Apoio")


Porém achei mais complexo analisar dessa forma, pois são muitas variaveis. Então decidi agrupar um pouco mais.

```python
# --- Conta quantos apoios cada aluno recebeu ---
df['TotalApoios'] = (
    (df['ApoioFamilia'] == 'yes').astype(int) +
    (df['ApoioEscola'] == 'yes').astype(int) +
    (df['ApoioExtra'] == 'yes').astype(int)
)

# --- Criar uma nova coluna com o total de apoios por aluno ---
def rotulo_apoio(n):
    if n == 0:
        return 'Nenhum apoio'
    elif n == 1:
        return '1 apoio'
    elif n == 2:
        return '2 apoios'
    else:
        return '3 apoios'

df['GrupoApoio'] = df['TotalApoios'].apply(rotulo_apoio)

ordem_apoios = ['Nenhum apoio', '1 apoio', '2 apoios', '3 apoios']

# --- Criação da figura ---
fig, ax = plt.subplots(figsize=(8, 5))

# --- Boxplot por GrupoApoio para ver a distribuição ---
sns.boxplot(x='GrupoApoio', y='P3', data=df, order=ordem_apoios)
configurar_eixos(ax, titulo='Distribuição das Notas (P3) por Número de Apoios')

# --- Plotagem ---
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.15)
plt.show()
```

- Nenhum Apoio
  - Mediana mais alta do que os demais grupos
  - Distribuição ampla, mas com alguns outliers abaixo de 4
  - Indica que alguns alunos sem apoio ainda conseguem desempenho razoável
- Um Apoio / Dois Apoios
  - Medianas levemente menores do que o grupo sem apoio
  - Distribuições bem parecidas entre si, com dispersão moderada
- Três Apoios
  - Mediana mais baixa do gráfico
  - Menor dispersão geral — indica consistência, mas em notas mais baixas

Apesar do senso comum apontar que mais apoio, melhor desempenho, esse gráfico sugere algo contraintuitivo:

Alunos com 3 apoios educacionais apresentam, em média, desempenho mais baixo do que os demais grupos — inclusive do grupo sem apoio.







## Violino

Esse gráfico em formato de violino (violinplot) traz uma camada muito interessante à análise de como o número de apoios educacionais influencia a nota final (P3). Vamos destrinchar os principais achados visuais:


```python
# --- Criação da figura ---
fig, ax = plt.subplots(figsize=(8, 5))

# --- Criação do gráfico ---
sns.violinplot(
    x='GrupoApoio',
    y='P3',
    data=df,
    order=ordem_apoios,
    inner='quartile',
    hue='GrupoApoio',        # especifica que o agrupamento é pelo x
    legend=False,            # evita legenda repetida
    palette='pastel'
)
configurar_eixos(ax, titulo='Distribuição da Nota (P3) por Número de Apoios — Violinplot')

# --- Plotagem ---
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.15)
plt.show()
```

- Nenhum Apoio
    - A distribuição é bem concentrada entre 10 e 15
    - Mediana por volta de 12
    ➤ Indica que alunos sem apoio conseguem desempenho mediano, estável e com menos extremos
- 1 Apoio
    - Muito semelhante ao grupo anterior, mas com leve achatamento da curva
    ➤ Isso sugere que ter apenas um tipo de apoio não melhora muito o desempenho
- 2 Apoios
    - A distribuição fica mais larga, com notas que se estendem de 5 a 20
    - Mediana levemente mais alta
    ➤ Mostra mais variabilidade: alguns se beneficiam muito, outros nem tanto
- 3 Apoios
    - Densidade comprimida entre 10 e 15, com menos extremos
    - Mediana ligeiramente abaixo das demais
    ➤ Indica um grupo com resultados consistentes, porém ligeiramente mais baixos

Ter 2 apoios parece ser o ponto de maior variação e potencial de melhoria, mesmo que não seja consistente.

Ter 3 apoios não se traduz em desempenho superior — pode estar associado a alunos em maior vulnerabilidade sendo alvo de múltiplas intervenções.

Nenhum apoio ou apenas um? O desempenho não é catastrófico, mas há uma limitação evidente de avanço.



# Mapa de Calor (heatmap)
 Mapa de calor entre número de apoios e tempo de estudo. A ideia é analisar se quem recebe mais apoio também estuda mais — ou não.



```python
# --- Criação da figura ---
fig, ax = plt.subplots(figsize=(8, 5))

# --- Criação do gráfico ---
heatmap_data = pd.crosstab(df['TotalApoios'], df['TempoEstudoDescricao'])
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='Blues')
plt.title('Frequência de Alunos por Nº de Apoios e Tempo de Estudo')
plt.xlabel('')
plt.ylabel('Nº de Apoios')

# --- Plotagem ---
plt.tight_layout()
plt.subplots_adjust(wspace=0.1, hspace=0.15)
plt.show()
```

- A maioria dos alunos tem 1 apoio e estuda entre 2 e 5 horas
  - Pico absoluto: 244 alunos com 1 apoio e 2 a 5 horas de estudo semanais
  - Isso mostra que esse é o perfil mais comum — apoio isolado (possivelmente escolar) + tempo de estudo moderado

- Alunos sem apoio ainda são muito numerosos
  - Total de alunos sem apoio: valores altos em todas as faixas
  - Especialmente em <2h (142 alunos) e 2 a 5h (128 alunos)
  - Ou seja, muitos estudantes estão em situação de vulnerabilidade dupla: pouco apoio e pouco estudo

- O grupo com 3 apoios é minúsculo
  - Apenas 23 alunos no total entre todas as faixas
  - Isso reforça que receber todos os tipos de apoio ao mesmo tempo é muito raro
  - Isso ajuda a explicar por que o desempenho desse grupo foi tão modesto: pequeno n, alta heterogeneidade

- Não há indício claro de que quem recebe mais apoio estuda mais
  - Ex: a faixa >10h/semana é pequena em todos os níveis de apoio
  - Mesmo com 3 apoios, apenas 4 alunos estudam mais de 10h
  - Ou seja: apoio educacional formal nem sempre se traduz em maior tempo dedicado ao estudo

É possível que o apoio esteja sendo direcionado a alunos com menor autonomia ou dificuldades crônicas

# Conclusão:




Os dados indicam que **alunos que estudam mais tendem a ter notas mais altas**, mas esse efeito é **moderado, gradual e não linear.** A média das notas cresce conforme o tempo de estudo aumenta, especialmente até a faixa de **5 a 10 horas semanais**, onde o ganho é mais evidente. No entanto, após esse ponto, o desempenho **se estabiliza ou até recua levemente**, sugerindo a existência de um limite prático no quanto estudar faz diferença.

Além disso, o tempo de estudo parece **reduzir o risco de notas muito baixas e aumentar a consistência no desempenho**, mesmo que não garanta excelência por si só. A presença de outliers em todas as faixas reforça **que outros fatores também influenciam fortemente o resultado final.**

Quanto ao apoio educacional, os resultados foram ainda mais surpreendentes. **Receber algum tipo de apoio (familiar, escolar ou externo) não se traduziu, em média, em notas mais altas.** O grupo sem apoio apresentou desempenho semelhante — e em alguns casos superior — ao de alunos com apoio isolado. Mesmo a combinação de múltiplos apoios (2 ou 3 tipos) **não garantiu melhor desempenho**, sendo que o grupo com 3 apoios teve, inclusive, a mediana mais baixa entre todos.

Isso sugere que o apoio educacional, da forma como está sendo oferecido, **não está sendo suficiente para transformar o desempenho acadêmico**. Além disso, **a maioria dos alunos com apoio ainda estuda pouco**, o que indica que o apoio não está mudando significativamente o comportamento de estudo.

Em resumo:
- Estudar mais ajuda, mas até certo ponto. O ganho é real, porém limitado.
- Apoio educacional, isoladamente, não garante melhor desempenho — e pode estar sendo direcionado a alunos com maiores dificuldades.
- Tempo de estudo e apoio são importantes, mas não suficientes. O desempenho acadêmico é multifatorial e exige uma abordagem integrada, que considere também motivação, ambiente, qualidade do ensino e fatores socioeconômicos.
